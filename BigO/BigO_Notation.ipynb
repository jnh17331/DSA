```python
##O(1)
def multiply_numbers(x):
    return x*x

print(multiply_numbers(206))
```

```python
##O(n)
def print_item(x):
    for i in range(x):
        print(i)

print(print_item(100))
```

```python
## Time Complexity O(n^2)

def print_item(x):
    for i in range(x):
        for j in range(x):
            for k in range(x):
                print(i, j, k)
    
        

print_item(10)
```

```python
## Space Complexity Practice

def pair_sum_seq(n):
    total = 0
    for i in range(n):
        total = total + pair_sum(i, i+1)
    return total

def pair_sum(a, b):
    return a+b

pair_sum_seq(4)
```

```python
## Big O(log n) - Binary Search

data = [1, 2, 3, 5, 7, 8, 9, 12, 125, 125126, 126161, 1216758, 46584, 5689, 35735, 357, 8576, 35737, 132, 76, 97, 35, 876, 487]

def binary_search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')


print(binary_search(data, 126161))
```
